```python
import pandas as pd

df = pd.read_csv("vizag_house_prices.csv")
print(df.head())
```

```python
df_encoded = pd.get_dummies(df , columns=["Locality"])

print(df_encoded.head())
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

x = df_encoded.drop("Price_Lakhs", axis=1)
y = df_encoded["Price_Lakhs"]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)

model=LinearRegression()
model.fit(x_train,y_train)
```

```python
# Step 1: Create input for new house
new_house = {
    'BHK': 2,
    'Size_sqft': 1600,

    # keep '1' one for which the area price u wanna know
    'Locality_Akkayyapalem': 0,
    'Locality_Bheemunipatnam': 0,
    'Locality_Dwaraka Nagar': 0,
    'Locality_Gajuwaka': 0,
    'Locality_Gopalapatnam': 0,
    'Locality_Kapuluppada': 0,
    'Locality_Madhurawada': 1,
    'Locality_NAD Junction': 0,
    'Locality_PM Palem': 0,
    'Locality_Seethammadhara': 0,
    'Locality_Yendada': 1
}

import pandas as pd

# Step 2: Convert to DataFrame
input_df = pd.DataFrame([new_house])

# Step 3: Predict
predicted_price = model.predict(input_df)

# Step 4: Show result
print(f"üè† Predicted Price: ‚Çπ{predicted_price[0]:.2f} Lakhs")
```
